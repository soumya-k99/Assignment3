```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
plt.rcParams.update({'font.size': 15})

z=np.linspace(0,50,150,endpoint=True)
#print(z)

k=4304.16
omega_m_0=[1.0, 0.7, 0.5, 0.3, 0]
omega_lam=[0,0.3, 0.5, 0.7, 1.0]
def E(z,p,q):

    e = omega_m_0[p] * ((1 + z) ** 3) + omega_lam[q]
    return 1/(np.sqrt(e))


da1=np.zeros(len(z))
da2=np.zeros(len(z))
da3=np.zeros(len(z))
da4=np.zeros(len(z))
da5=np.zeros(len(z))

for i in range(len(z)):
    da1[i]=integrate.quadrature(E,0,z[i],args=(0,0))[0]*(1/(1+z[i]))*k
    da2[i]=integrate.quadrature(E,0,z[i],args=(1,1))[0]*(1/(1+z[i]))*k
    da3[i]=integrate.quadrature(E,0,z[i],args=(2,2))[0]*(1/(1+z[i]))*k
    da4[i]=integrate.quadrature(E,0,z[i],args=(3,3))[0]*(1/(1+z[i]))*k
    da5[i]=integrate.quadrature(E,0,z[i],args=(4,4))[0]*(1/(1+z[i]))*k
#print(da1)
fig=plt.figure(figsize=(9,9))
plt.plot(z,da1  , label='Model 1',lw='4')
plt.plot(z,da2  , 'Turquoise',label='Model 2',lw='4')
plt.plot(z,da3  , label='Model 3',lw='4')
plt.plot(z,da4  ,'chartreuse', label='Model 4',lw='4')
plt.plot(z,da5 , 'darkmagenta',label='Model 5',lw='4')
plt.xlabel('Redshift(z)')
plt.ylabel('Angular diameter Distance   ('+r'$d_A$'+' in Mpc )')
plt.legend()
plt.title('Angular diameter Distance as a function of Redshift',fontweight="bold")
plt.grid()
plt.show
plt.savefig('da_1.png')





```
